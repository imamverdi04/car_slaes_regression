```python
chip_names = [
    "Kettle","Natural Chip Compny", "Cobs","Doritos", "Tyrrells Crisps", "Tostitos", "Infuzions", "Smiths", "Thins",
    "Dorito", "Pringles", "Twisties", "Grain Waves", "Cheezels", "Old El Paso",
    "Natural Chip Co", "Natural Chip Compny", "Natural ChipCo", "NCC", "CCs",
    "Red Rock Deli", "RRD", "Woolworths", "WW", "Cheetos", "Sunbites", "Snbts",
    "French Fries", "Burger Rings", "Infzns", "Smith", "GrnWves", "Dorito", "Natural Chip Compny",
]
```

```python
def extract_chip_name(row, chip_names):
    normalized_row = re.sub(r'\s+', ' ', row.strip().lower())

    for chip in chip_names:
        normalized_chip = re.sub(r'\s+', ' ', chip.strip().lower())
        if normalized_chip in normalized_row:
            row = re.sub(r'\b' + re.escape(chip) + r'\b', '', row, flags=re.IGNORECASE).strip()
            return chip, row

    return None, row

df['chip_name'], df['Məhsul_adi'] = zip(*df['Məhsul_adi'].apply(extract_chip_name, args=(chip_names,)))
```

```python
def handle_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df[column] = df[column].apply(lambda x: lower_bound if x < lower_bound else upper_bound if x > upper_bound else x)

handle_outliers(df, 'weight')
handle_outliers(df, 'Ümumi satış')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

def select_numeric_categorical_columns(df):
    numeric_features = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
    categorical_features = df.select_dtypes(exclude=['int64', 'float64']).columns.tolist()
    return numeric_features, categorical_features

X_train, X_test, y_train, y_test = train_test_split(
    df.drop('Ümumi satış', axis=1),
    df['Ümumi satış'],
    test_size=0.2,
    random_state=42
)

numeric_features, categorical_features = select_numeric_categorical_columns(X_train)

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

X_train_transformed = preprocessor.fit_transform(X_train)
X_test_transformed = preprocessor.transform(X_test)
```
